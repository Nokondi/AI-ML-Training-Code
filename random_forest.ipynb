# Random Forest

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

# Load the breast dataset
data = load_breast_cancer()
breast_cancer = pd.DataFrame(data.data, columns=data.feature_names)

# Just use only 2 features for demonstration
breast_cancer = breast_cancer.loc[:, ["mean radius", "mean compactness"]]

display(breast_cancer)
```

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
X = StandardScaler().fit_transform(breast_cancer)
plt.scatter(X[:, 0], X[:, 1], c=data.target)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Set oob_score to true to see the self validation score
model_tree = []
tree_nums = [200, 400, 600, 1000, 10000]

for tree_num in tree_nums:
    model = RandomForestClassifier(oob_score=True, n_estimators=tree_num, random_state=1).fit(X, data.target)
    model_tree.append(model)
```

```python
import numpy as np

# Plot the decision boundry

# Get the grid value
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))


# Plot the figure

plt.figure(figsize=(12, 60))

for index, model in enumerate(model_tree):
    plt.subplot(len(model_tree), 1, index + 1)
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.title(f"Decision boundry of Random Forest with {tree_nums[index]} trees")
    # Plot contour with color filled
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)

    # Plot training data
    plt.scatter(X[:, 0], X[:, 1], c=data.target)
```

```python
# See the oob score of the tree
for index, model in enumerate(model_tree):
    print(f'The oob score of {tree_nums[index]} trees is {model.oob_score_}')
```
