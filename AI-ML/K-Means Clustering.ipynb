#Clustering Countries by using Unsupervised Learning for HELP International
##Objective:
To categorise countries using socio-economic and health factors that determine the overall development of the country.

##About organization:
HELP International is an international humanitarian NGO that is committed to fighting poverty and providing the people of impoverished countries with aid and disaster relief.

##Problem Statement:
HELP International has been able to raise around $ 10 million. Now the CEO of the NGO needs to decide how to use this money strategically and effectively. The CEO wants to choose the countries that are in the direst need of aid. Your job as a data scientist is to categorise the countries using some socio-economic and health factors that determine the overall development of the country; then, you need to suggest the countries which the CEO needs to focus on the most.
##Data Dictionary:
**country** (text): name of the country

**child_mort** (float): Deaths of children under 5 years of age per 1000 live births

**exports** (float): Exports of goods and services per capita. Given as % of the GDP per capita

**health** (float): Total health spending per capita. Given as %age of GDP per capita

**imports** (float): Imports of goods and services per capita. Given as %age of the GDP per capita

**income** (float): Net income per person

**inflation** (float): The measurement of the annual growth rate of the Total GDP

**life_expec** (float): Average life expectancy of citizens in the country

**total_fer** (float): Average number of children born in each family

**gdpp** (float): The GDP per capita. Calculated as the Total GDP divided by the total population.

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

## Load data

```python
url = "https://drive.google.com/uc?export=download&id=1wAMxE_xWS01MM5nbjEmgRRH6hXea3ft_"

data = pd.read_csv("https://drive.google.com/uc?export=download&id=1wAMxE_xWS01MM5nbjEmgRRH6hXea3ft_")
country = pd.DataFrame(data['country'])
stats = data.loc[:, 'child_mort':'gdpp']
stats.head()
```

##Normalize

```python
for column in stats.columns:
    stats[column] = MinMaxScaler().fit_transform(np.array(stats[column]).reshape(-1, 1))

stats.head()
```

##Preliminary Visual Analysis

```python
p = sns.PairGrid(stats)
p.map_diag(sns.histplot)
p.map_offdiag(sns.scatterplot)
```

##Clustering

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(stats)
predict = kmeans.predict(stats)
stats['class'] = predict
stats.head()
```

##Visualize Clusters

```python
p = sns.PairGrid(stats, hue='class', palette='viridis')
p.map_diag(sns.histplot)
p.map_offdiag(sns.scatterplot)
#handles = list(p._legend_data.values())
#labels = {}
#for i, n in enumerate(iris.target_names):
#  labels[n] = handles[i]

p.add_legend(legend_data=p._legend_data, title='class')
```

```python
data['class'] = np.array(stats['class'])
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(data.sort_values(by='class'))
```
