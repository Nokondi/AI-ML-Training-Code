```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-whitegrid')
plt.style.available
```

```python
x = np.linspace(0, 10, 1000)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
fig.savefig('sin_cos.svg')
fig.canvas.get_supported_filetypes()
```

```python
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))
plt.subplot(2, 2, 4)
plt.plot(x, np.tanh(x))
```

```python
fig, axes = plt.subplots(2, 2)
for i, ax in enumerate(axes.flat):
  ax.plot(x, np.sin(x))
```

```python
c = ['blue', '#09814a', 'chartreuse', '#ABCD12', 'yellow', '0.6']
l = ['solid', 'dotted', 'dashed', 'dashdot', ':', '-.']
for i in range(0, 6, 1):
  plt.plot(x, np.sin(x - i), color=c[i], linestyle=l[i])
```

```python
plt.plot(x, np.sin(x), color=(0.1, 0.8, 0.7))
plt.ylim(-1.25, 1.25)
plt.xlabel('time')
plt.ylabel('intensity')
plt.title('sine wave')
```

```python
x = np.linspace(0, 10, 1000)
fig = plt.figure()
plt.plot(x, np.sin(x), '-', label='sin x', color='#7ee8fa')
plt.plot(x, np.cos(x), '--', label='cos x', color='#eec0c6')
plt.ylim(-1.25, 1.25)
plt.legend()
```

```python
x = np.linspace(0, 10, 20)
y = np.sin(x)
plt.plot(x, y, 'o', color='#80ff72')
plt.plot(x, y, color='#E58C8A')
plt.ylim(-1.25, 1.25)
```

```python
rng = np.random.RandomState(0)
markers = ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']
for m in markers:
  x = rng.rand(5)
  y = rng.rand(5)
  plt.plot(x, y, m, label='marker={0}'.format(m))
  # plt.plot(x, y)
plt.legend()
plt.xlim(0, 1.5)
```

```python
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.4, cmap='viridis')
plt.colorbar()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], c=iris.target, s=200*features[3], cmap='viridis', alpha=0.3)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
print(iris.feature_names[3])
```
