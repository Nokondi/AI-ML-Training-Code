```python
import plotly.express as px
import pandas as pd
import requests
import openpyxl
from itertools import islice

wbdl = requests.get('https://advocacy.code.org/2021_State_of_CS_data.xlsx')
data_file = open('data.xlsx', 'wb')
data_file.write(wbdl.content)
data_file.close()
workbook = openpyxl.load_workbook('data.xlsx')
sheet_list = workbook.sheetnames
i = 0
dataframes = {}
for ws in workbook:
  data = ws.values
  ws.delete_rows(0, 2)
  cols = ['states']
  cols += list(next(data)[1:])
  data = list(data)
  idx = [r[0] for r in data if r[0] != None and r[0][0] != '*']
  data = (r for r in data if r[0] != None and r[0][0] != '*')
  df = pd.DataFrame(data, columns=cols)
  dataframes[sheet_list[i]] = df
  i += 1
```

```python
percent = []
states = []
dataframes['AccessReport']
for i in range(0, len(dataframes['AccessReport']['2020–21'])):
  row = dataframes['AccessReport']['2020–21'][i]
  states.append(dataframes['AccessReport']['states'][i].replace('*', ''))
  if not pd.isna(row) and type(row) is not str:
    percent.append(float(row))
  else:
    row = dataframes['AccessReport']['2019–20'][i]
    if not pd.isna(row) and type(row) is not str:
      percent.append(float(row))
    else:
      row = dataframes['AccessReport']['2018–19'][i]
      if not pd.isna(row) and type(row) is not str:
        percent.append(float(row))
      else:
        row = dataframes['AccessReport']['2017–18'][i]
        if not pd.isna(row) and type(row) is not str:
          percent.append(float(row))
        else:
          row = dataframes['AccessReport']['2016–17'][i]
          if not pd.isna(row) and type(row) is not str:
            percent.append(float(row))

for i in range(0, len(percent)):
  print(states[i], percent[i])

fig = px.choropleth(dataframes['AccessReport'], locations=states, locationmode="USA-states", color=percent, scope="usa", color_continuous_scale=["#8A84D6", "#4D4084"])
fig.update_geos(visible=False)
# fig.update_traces(autocolorscale=False, colorscale=[[1.0, 'rgb(77, 64, 132)'], [0, 'rgb(138, 132, 214)']])
fig.add_scattergeo(locations=states, connectgaps=False, text=percent, mode="text", textfont={"color":"white", "family":"Open Sans", "size": 18}, locationmode="USA-states")
fig.show()
```
