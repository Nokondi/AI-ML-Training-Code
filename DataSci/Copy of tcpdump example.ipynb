```python
import subprocess as sub
import time
import geocoder

# Open subprocess and run tcpdump from the command line
p = sub.Popen(('tcpdump', '-l', '-A'), stdout=sub.PIPE)

# Wait 10s
time.sleep(10)

# End task
p.terminate()

# Process and store packets
packets = []
for row in iter(p.stdout.readline, b''):
    packets.append(str(row, 'utf-8').replace('\n', ''))

# View / analyze packets
ips = []
for pack in packets:
    print(pack)
    if 'IP6' in pack:
        raw_ips = pack.split('IP6 ')[1].split(': ')[0].split(' > ')
        for i in raw_ips:
            ips.append(i.split('.')[0])
    elif 'IP' in pack:
        raw_ips = pack.split('IP ')[1].split(': ')[0].replace('.https', '').split(' > ')
        for i in raw_ips:
            ips.append(i)

# Print location info for ips
for ip in ips:
    ip = geocoder.ip(ip)
    print(ip.latlng, ip.city)
```
