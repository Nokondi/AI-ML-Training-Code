```python
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set()
sns.set_style('darkgrid')
```

```python
iris = load_iris()
iris.feature_names.append('class')
data = np.column_stack((iris.data, iris.target))
iris_pd = pd.DataFrame(data=data, columns=iris.feature_names)
print(iris_pd)
```

```python
# p = sns.pairplot(iris_pd, hue='class', palette='viridis')
p = sns.PairGrid(iris_pd, hue='class', palette='viridis')
p.map_diag(sns.histplot)
p.map_offdiag(sns.scatterplot)
handles = list(p._legend_data.values())
labels = {}
for i, n in enumerate(iris.target_names):
  labels[n] = handles[i]

p.add_legend(legend_data=labels, title='class', label_order=iris.target_names, )
```

```python
train_data, test_data, train_class, test_class = train_test_split(iris.data, iris.target)
train_data = tf.constant(train_data)
train_class = tf.constant(train_class)
test_data = tf.constant(test_data)
test_class = tf.constant(test_class)

model = tf.keras.models.Sequential([tf.keras.layers.Input(shape=(4,)), 
                                    tf.keras.layers.Dense(100, activation='relu'),
                                    tf.keras.layers.Dropout(0.2),
                                    tf.keras.layers.Dense(3, activation=tf.keras.activations.softmax)])

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()

model.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])

history = model.fit(train_data, train_class, epochs=100)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].plot(history.history['accuracy'])
ax[0].set_title('Accuracy')
ax[1].plot(history.history['loss'])
ax[1].set_title('Loss')
```

```python
model.evaluate(test_data, test_class)
p = model.predict(test_data)
p = tf.constant(p)
p = np.round(p.numpy(), 3)
p = tf.math.argmax(p, axis=1)
print(classification_report(test_class, p, target_names=iris.target_names))
c = tf.math.confusion_matrix(test_class, p)
plt.figure(figsize=(10,10))
ax = sns.heatmap(c, square=True, annot=True, fmt='d', cbar=False, xticklabels=iris.target_names, yticklabels=iris.target_names)
ax.set_xlabel('True')
ax.set_ylabel('Predicted')
```
